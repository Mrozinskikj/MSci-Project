```python
def getDistance(tHist, rHist, interval):
    A = tHist / sum(tHist)
    B = rHist / sum(rHist)
    
    if(len(A)>len(B)):
        B = np.pad(B,(0,len(A)-len(B)))
    if(len(B)>len(A)):
        A = np.pad(A,(0,len(B)-len(A)))
    
    n = len(A)
    dist = np.zeros(n)
    for x in range(n-1):
        dist[x+1] = A[x]-B[x]+dist[x]
    wasserstein = np.sum(abs(dist))
    
    wasserstein = wasserstein * interval

    return wasserstein
```
